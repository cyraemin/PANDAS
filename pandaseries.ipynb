```python
#panda series
import numpy as np
import pandas as pd
```

```python
#to create series
#series are one dimensional labelled array capable of holding any data type
list=[1,2,3]
arr=np.array([4,5,6])
labels=['a','b','c']
dict={1:20,2:40,3:60}
```

```python
pan=pd.Series(list)
pan
```

```python
pan=pd.Series(arr)
pan
```

```python
pan=pd.Series(dict)
pan
```

```python
pan=pd.Series(list,index=labels)
pan
```
