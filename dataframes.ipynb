```python
import numpy as np
import pandas as pd
#when you combine series together,that is referred to as dataframes
```

```python
#1.creating dataframe
data = {
    'Name': ['John', 'Anna', 'Peter', 'Linda'],
    'Age': [28, 34, 29, 42],
    'City': ['New York', 'Paris', 'Berlin', 'London'],
    'Salary': [65000, 70000, 62000, 85000]
}
df=pd.DataFrame(data)
df
```

```python
data_list = [
    ['John', 28, 'New York', 65000],
    ['Anna', 34, 'Paris', 70000],
    ['Peter', 29, 'Berlin', 62000],
    ['Linda', 42, 'London', 85000]
]
df2=pd.DataFrame(data_list)
df2
columns=["Name","Age","City","Salary"]
df2=pd.DataFrame(data_list,columns=columns)
df2
```

```python
#2.selection and indexing of columns
#to select one col-
df2[["Age"]]
```

```python
#to select more than one columns,make list of list
df2[["Age","City"]]
```

```python
#making new col-
df2["Food"]=["Apple","Banana","Mango","Orange"]
df2
```

```python
#3. remove a col or more than 1 col
df2.drop(["Food"],axis=1,inplace=True)#use the inplace keyword,otherwise only copy is returned and not actually deleted
df2
```

```python
#4.selecting row
df2.loc[[0,1]]
```

```python
#selecting subset of rows and columns
df2.loc[[0,1]][["City","Salary"]]
```

```python
#conditional selection
df2[(df2["Age"]>30)&(df2["City"]=='Paris')]
```
