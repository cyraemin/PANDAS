```python
import numpy as np
import pandas as pd
#missing data
```

```python
data = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [1, 2, 3, 4, 5],
    'C': [1, 2, 3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5]
}
df=pd.DataFrame(data)
df
```

```python
#finding missing data- based on columns
df.isna()
```

```python
df.isna().sum()#how many missing data in each column
```

```python
df.isna().any()#checks whether any column has no missing data or no
```

```python
#removing missing data- works on basis of rows
df.dropna()
```

```python
df.dropna(thresh=3)#returns rows having at least three non null values
```

```python
#filling missing data- works on the basis of columns
df.fillna(0)
```

```python
values={'A':1,'B':2,'C':3,'D':4}
df.fillna(value=values)
```
